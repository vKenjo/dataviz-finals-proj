# 1. Project Background

## The Global Poverty Challenge

Extreme poverty remains one of humanity's most pressing challenges. Despite remarkable economic growth and technological advancement, approximately **700 million people**—nearly 9% of the global population—still live on less than $2.15 per day as of 2023 (World Bank, 2023).

## UN Sustainable Development Goal 1: No Poverty

The United Nations' Sustainable Development Goal 1 aims to **"end poverty in all its forms everywhere" by 2030**. This ambitious target encompasses:

- Eradicating extreme poverty (living on <$2.15/day)
- Reducing poverty by half according to national definitions
- Implementing social protection systems
- Ensuring equal rights to economic resources and basic services

## Historical Context

The past three decades have witnessed unprecedented poverty reduction:
- **1990:** ~36% of the global population lived in extreme poverty
- **2023:** ~9% live in extreme poverty
- **Result:** Over 1.2 billion people lifted out of poverty

However, progress has been uneven:
- **East Asia:** Reduced from 66% to <1% (driven by China)
- **Sub-Saharan Africa:** Remains above 40% in many countries

## Project Significance

This project contributes by:
1. Providing comprehensive visual analysis of 33 years of global poverty data
2. Identifying correlations between poverty and development indicators
3. Comparing successful and challenging cases
4. Generating evidence-based policy recommendations

## Limitations

1. **Data Gaps**: 
   - Not all countries have continuous annual data
   - Philippines has gaps in certain years for some indicators
   - Some SEA countries lack consistent poverty measurement

2. **Measurement Consistency**: 
   - Poverty measurement methodologies vary across countries
   - PPP (Purchasing Power Parity) calculations updated periodically
   - International poverty line ($2.15/day) may not reflect local realities

3. **Sub-national Variation**: 
   - Country-level data masks regional disparities within countries
   - Philippines has significant urban-rural and regional poverty gaps
   - Island geography creates unique challenges not captured in national data

4. **Recent Events**: 
   - COVID-19 impact (2020-2023) may not be fully reflected in latest data
   - Some countries have reporting lags of 1-2 years

5. **Causation vs. Correlation**: 
   - Statistical correlations don't necessarily prove causation
   - Multiple factors interact in complex ways to affect poverty

6. **Comparative Limitations**:
   - Philippines' unique geography (archipelago) and demographics differ from mainland SEA countries
   - Historical, political, and institutional contexts vary significantly across countries

## 5b. Data Preparation

### Step 1: Import Libraries and Setup

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
from pathlib import Path
from datetime import datetime

warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

# Create directories
FIGURES_DIR = Path('figures')
DATA_DIR = Path('data')
FIGURES_DIR.mkdir(exist_ok=True)
DATA_DIR.mkdir(exist_ok=True)

print("✓ Libraries imported successfully")
print(f"✓ Figures will be saved to: {FIGURES_DIR}")
print(f"✓ Data directory: {DATA_DIR}")
```

### Step 2: Load or Download Data

We'll try to load existing data first, otherwise we'll need to download from World Bank API.

```python
# Check if data file exists
data_file = DATA_DIR / 'data.csv'

if data_file.exists():
    print(f"Loading existing data from {data_file}...")
    df = pd.read_csv(data_file)
    print(f"✓ Loaded {len(df):,} records")
    print(f"✓ Columns: {list(df.columns)}")
else:
    print("No existing data file found.")

# Display basic info
print("\nDataset Overview:")
print(f"Shape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

### Step 3: Data Cleaning and Preprocessing

```python
# Create a copy for processing
df_clean = df.copy()

print("Data Cleaning Steps:")
print("=" * 50)

# 1. Filter out non-country aggregates and groups
print("\n1. Filtering Non-Country Entries:")
print(f"   Original unique entities: {df_clean['country'].nunique()}")

# Define patterns to exclude (World Bank aggregates, income groups, regional groups)
exclude_patterns = [
    'World', 'IBRD', 'IDA', 'income', 'dividend', 'Heavily indebted',
    'Fragile', 'OECD', 'Euro area', 'European Union', 'Arab World',
    'Central Europe', 'Caribbean', 'East Asia', 'Latin America',
    'Middle East', 'North Africa', 'North America', 'Pacific island',
    'South Asia', 'Sub-Saharan', 'West Bank', 'Small states',
    'Least developed', 'Low income', 'Lower middle', 'Upper middle',
    'High income', 'Post-demographic', 'Pre-demographic', 'Late-demographic',
    'Early-demographic', 'Not classified', 'total', 'only', '&'
]

# Create a mask for rows to keep (actual countries)
mask = ~df_clean['country'].str.contains('|'.join(exclude_patterns), case=False, na=False)
df_clean = df_clean[mask].copy()

print(f"   After filtering: {df_clean['country'].nunique()} unique countries")
print(f"   Removed {df['country'].nunique() - df_clean['country'].nunique()} non-country entries")

# 2. Check for missing values
print("\n2. Missing Value Analysis:")
missing_pct = (df_clean.isnull().sum() / len(df_clean) * 100).round(2)
missing_df = pd.DataFrame({
    'Column': missing_pct.index,
    'Missing %': missing_pct.values
}).sort_values('Missing %', ascending=False)
print(missing_df[missing_df['Missing %'] > 0])

# 3. Data type conversions
print("\n3. Data Types:")
print(df_clean.dtypes)

# 4. Remove duplicates
before_dedup = len(df_clean)
df_clean = df_clean.drop_duplicates()
after_dedup = len(df_clean)
print(f"\n4. Removed {before_dedup - after_dedup} duplicate rows")

# 5. Validate country count
print(f"\n5. Final country count: {df_clean['country'].nunique()} countries")
print(f"   Total observations: {len(df_clean)}")

# 6. Summary statistics
print("\n6. Summary Statistics:")
df_clean.describe()
```

## 5c. Exploratory Data Analysis (EDA)

**Goal**: Understand the data distribution, identify patterns, and detect outliers through preliminary visualizations.

We will create **3 exploratory visualizations**:

```python
# EDA 1: Distribution Analysis - Four separate visualizations

# Define ASEAN countries (all 11 members)
sea_countries = [
    'Philippines', 'Vietnam', 'Thailand', 'Indonesia', 'Malaysia', 
    'Singapore', 'Myanmar', 'Cambodia', 'Lao PDR', 'Brunei Darussalam', 
    'Timor-Leste'
]

# Prepare data for latest available year per country
latest_available = df_clean.sort_values('year').groupby('country').last().reset_index()

# Define indicators for analysis
indicators = ['poverty_headcount', 'gdp_per_capita', 'life_expectancy', 'electricity_access']

# Create color scheme
color_phl = '#d62728'  # Red for Philippines
color_sea = '#1f77b4'  # Blue for other SEA
color_global = '#7f7f7f'  # Gray for global

# Create 4 separate distribution plots
for indicator in indicators:
    plt.figure(figsize=(12, 6))
    
    # Filter out missing values
    data_global = latest_available[indicator].dropna()
    data_sea = latest_available[latest_available['country'].isin(sea_countries)][indicator].dropna()
    data_phl = latest_available[latest_available['country'] == 'Philippines'][indicator].dropna()
    
    # Plot distributions
    if len(data_global) > 0:
        plt.hist(data_global, bins=30, alpha=0.5, label='Global', color=color_global, edgecolor='black')
    if len(data_sea) > 0:
        plt.hist(data_sea, bins=15, alpha=0.7, label='ASEAN Countries', color=color_sea, edgecolor='black')
    if len(data_phl) > 0:
        plt.axvline(data_phl.values[0], color=color_phl, linestyle='--', linewidth=2, label='Philippines')
    
    plt.xlabel(indicator.replace('_', ' ').title(), fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title(f'Distribution of {indicator.replace("_", " ").title()}', fontsize=14, fontweight='bold')
    plt.legend(fontsize=10)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
print(f"\nNote: ASEAN countries analyzed: {len(sea_countries)} members")
print(f"Countries: {', '.join(sorted(sea_countries))}")
```

### EDA 2: Poverty Trends Heatmap - Temporal Patterns

```python
# EDA 2: Poverty Trends Heatmap - Limited to 15 countries with SEA/Philippines prioritized

# Create pivot table
poverty_by_year = df_clean.pivot_table(
    values='poverty_headcount', 
    index='country', 
    columns='year', 
    aggfunc='mean'
)

# Count data availability per country
data_count = poverty_by_year.notna().sum(axis=1)

# Separate ASEAN countries and others
sea_in_data = [c for c in sea_countries if c in poverty_by_year.index]
other_countries = [c for c in poverty_by_year.index if c not in sea_countries]

# Get top countries by data availability (excluding ASEAN)
other_sorted = data_count[other_countries].sort_values(ascending=False)

# Select: All ASEAN countries + Philippines + top others to reach 15 total
selected_countries = []
if 'Philippines' in poverty_by_year.index and 'Philippines' not in selected_countries:
    selected_countries.append('Philippines')
selected_countries.extend([c for c in sea_in_data if c != 'Philippines'])
remaining_slots = 15 - len(selected_countries)
if remaining_slots > 0:
    selected_countries.extend(other_sorted.head(remaining_slots).index.tolist())

# Limit to 15
selected_countries = selected_countries[:15]

# Create heatmap data
heatmap_data = poverty_by_year.loc[selected_countries].sort_values(by=poverty_by_year.columns[-1], ascending=False)

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, cmap='YlOrRd', cbar_kws={'label': 'Poverty Ratio (%)'}, 
            linewidths=0.5, linecolor='gray', annot=False)
plt.title('Poverty Ratio Trends Over Time (SEA Countries + Top Global)', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.tight_layout()
plt.show()

print(f"Showing {len(selected_countries)} countries:")
print(f"  ASEAN: {len([c for c in selected_countries if c in sea_countries])} countries")
print(f"  Others: {len([c for c in selected_countries if c not in sea_countries])} countries")
```

### EDA 3: Regional Comparison - Boxplots

```python
# EDA 3: Regional Comparison - Boxplot showing Global vs ASEAN vs Philippines

# Select key indicator for comparison
indicator = 'poverty_headcount'

plt.figure(figsize=(10, 7))

# Prepare data
global_data = latest_available[indicator].dropna()
sea_data = latest_available[latest_available['country'].isin(sea_countries)][indicator].dropna()
phl_data = latest_available[latest_available['country'] == 'Philippines'][indicator].dropna()

# Create boxplot data
data_to_plot = [global_data, sea_data, phl_data]
labels = ['Global\n(217 countries)', 'ASEAN\n(11 countries)', 'Philippines']
colors = ['#7f7f7f', '#1f77b4', '#d62728']

# Create vertical boxplot
bp = plt.boxplot(data_to_plot, labels=labels, patch_artist=True, vert=True,
                 showmeans=True, meanline=True, widths=0.6)

# Color the boxes
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

# Styling
for element in ['whiskers', 'fliers', 'means', 'medians', 'caps']:
    plt.setp(bp[element], color='black', linewidth=1.5)

plt.ylabel('Poverty Headcount (% at $2.15/day)', fontsize=12)
plt.title('Poverty Distribution Comparison\nGlobal vs ASEAN vs Philippines', 
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print(f"Boxplot shows: median (orange), mean (green), IQR (box), and outliers")
print(f"  Global median: {global_data.median():.1f}%")
print(f"  ASEAN median: {sea_data.median():.1f}%")
print(f"  Philippines: {phl_data.values[0]:.1f}%")
```

## 5d. Data Visualization

**Goal**: Create polished, informative visualizations to answer our research questions.

We will create **5 main visualizations** that tell the story of global poverty reduction:

```python
# Visualization 1: Interactive Choropleth Map

# Use latest available data for mapping
map_data = latest_available[['country', 'poverty_headcount', 'year']].copy()

# For countries without poverty data, use a proxy (inverse of life expectancy or GDP indication)
# This helps populate the map
map_data_filled = latest_available[['country', 'poverty_headcount', 'life_expectancy', 'gdp_per_capita', 'year']].copy()

# Create hover text with available data
map_data_filled['hover_text'] = map_data_filled.apply(
    lambda row: f"{row['country']}<br>Poverty: {row['poverty_headcount']:.1f}% ({row['year']:.0f})" 
    if pd.notna(row['poverty_headcount']) 
    else f"{row['country']}<br>No poverty data available", 
    axis=1
)

fig = px.choropleth(
    map_data_filled,
    locations='country',
    locationmode='country names',
    color='poverty_headcount',
    hover_name='country',
    hover_data={'poverty_headcount': ':.2f', 'year': ':.0f', 'country': False},
    color_continuous_scale='YlOrRd',
    labels={'poverty_headcount': 'Poverty Headcount (%)'},
    title='Global Poverty Headcount Distribution (Latest Available Year per Country)'
)

fig.update_geos(showcountries=True, countrycolor="lightgray")
fig.update_layout(height=600, width=1000)
fig.show()

print(f"Countries with poverty data: {map_data['poverty_headcount'].notna().sum()}")
print(f"Countries without poverty data: {map_data['poverty_headcount'].isna().sum()}")
```

### Visualization 2: Poverty Trends Over Time (ASEAN Focus)

```python
# VISUALIZATION 7: Sankey Diagram - Poverty Transitions with Context
print("Creating Sankey diagram for poverty transitions...")

def categorize_poverty(rate):
    if pd.isna(rate):
        return None
    elif rate < 3:
        return 'Very Low (<3%)'
    elif rate < 10:
        return 'Low (3-10%)'
    elif rate < 20:
        return 'Moderate (10-20%)'
    elif rate < 40:
        return 'High (20-40%)'
    else:
        return 'Very High (>40%)'

# Get poverty data for 1990 and 2020
poverty_1990 = df_clean[df_clean['year'].between(1988, 1992)].groupby('country')['poverty_headcount'].mean()
poverty_2020 = df_clean[df_clean['year'].between(2018, 2023)].groupby('country')['poverty_headcount'].mean()

# Find common countries
common_countries = list(set(poverty_1990.index) & set(poverty_2020.index))

# Create transitions dataframe
transitions = pd.DataFrame({
    'country': common_countries,
    'poverty_1990': [poverty_1990[c] for c in common_countries],
    'poverty_2020': [poverty_2020[c] for c in common_countries]
})

transitions['category_1990'] = transitions['poverty_1990'].apply(categorize_poverty)
transitions['category_2020'] = transitions['poverty_2020'].apply(categorize_poverty)
transitions = transitions.dropna()

# Check Philippines status for highlighting
phl_in_data = 'Philippines' in transitions['country'].values
if phl_in_data:
    phl_row = transitions[transitions['country'] == 'Philippines'].iloc[0]
    phl_context = f"Philippines: {phl_row['category_1990']} → {phl_row['category_2020']}"
else:
    phl_context = "Philippines: No data for this period"

transition_counts = transitions.groupby(['category_1990', 'category_2020']).size().reset_index(name='count')

if len(transition_counts) > 0:
    categories = ['Very Low (<3%)', 'Low (3-10%)', 'Moderate (10-20%)', 'High (20-40%)', 'Very High (>40%)']
    
    source_nodes = [f"{cat} (1990)" for cat in categories]
    target_nodes = [f"{cat} (2020)" for cat in categories]
    all_nodes = source_nodes + target_nodes
    
    # Create node mapping
    node_map = {node: idx for idx, node in enumerate(all_nodes)}
    
    # Prepare source, target, value lists
    sources = []
    targets = []
    values = []
    
    for _, row in transition_counts.iterrows():
        source_label = f"{row['category_1990']} (1990)"
        target_label = f"{row['category_2020']} (2020)"
        
        if source_label in node_map and target_label in node_map:
            sources.append(node_map[source_label])
            targets.append(node_map[target_label])
            values.append(row['count'])
    
    # Calculate statistics
    improved_count = (transitions['poverty_2020'] < transitions['poverty_1990']).sum()
    worsened_count = (transitions['poverty_2020'] > transitions['poverty_1990']).sum()
    avg_reduction = (transitions['poverty_1990'].mean() - transitions['poverty_2020'].mean())
    
    # Create Sankey diagram
    fig = go.Figure(data=[go.Sankey(
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color='black', width=0.5),
            label=all_nodes,
            color=['#2E86AB', '#54A0C8', '#83C5D3', '#A9D6E5', '#D9E9F1',
                   '#2E86AB', '#54A0C8', '#83C5D3', '#A9D6E5', '#D9E9F1']
        ),
        link=dict(
            source=sources,
            target=targets,
            value=values,
            color='rgba(0,0,0,0.2)'
        )
    )])
    
    fig.update_layout(
        title_text=f"Poverty Category Transitions (1990 → 2020)<br><sub>{improved_count} improved | {worsened_count} worsened | {phl_context}</sub>",
        title_font_size=16,
        height=600,
        font_size=10,
        annotations=[
            dict(
                text=f"<b>Context:</b> Total countries tracked: {len(transitions)} | "
                     f"Average poverty reduction: {avg_reduction:.1f} percentage points<br>"
                     f"Flow width represents number of countries transitioning between categories",
                x=0.5,
                y=-0.1,
                xref='paper',
                yref='paper',
                showarrow=False,
                font=dict(size=11, color='gray'),
                align='center'
            )
        ]
    )
    
    fig.write_html(FIGURES_DIR / 'viz_07_sankey.html')
    fig.show()
    
    print(f"✓ Sankey diagram created with {len(transitions)} countries")
    print(f"  Countries that improved: {improved_count}")
    print(f"  Countries that worsened: {worsened_count}")
    print(f"  Average poverty reduction: {avg_reduction:.1f} percentage points")
else:
    print("⚠ Not enough data for Sankey diagram")
```

```python
# Visualization 4: Multi-Dimensional Comparison - Philippines vs ASEAN Average

import numpy as np
from math import pi

# Define indicators for radar chart
indicators_radar = ['poverty_headcount', 'gdp_per_capita', 'life_expectancy', 
                   'electricity_access', 'primary_enrollment', 'secondary_enrollment']

# Get Philippines data
phl_latest = latest_available[latest_available['country'] == 'Philippines'][indicators_radar]

if len(phl_latest) == 0:
    print("⚠ Philippines data not found")
else:
    phl_latest = phl_latest.iloc[0]
    
    # Calculate ASEAN average
    sea_latest = latest_available[latest_available['country'].isin(sea_countries)][indicators_radar]
    sea_avg = sea_latest.mean()
    
    # Normalize values to 0-1 scale
    all_data = latest_available[indicators_radar]
    normalized_phl = []
    normalized_sea = []
    categories = []
    
    for indicator in indicators_radar:
        if pd.notna(phl_latest[indicator]) and pd.notna(sea_avg[indicator]):
            min_val = all_data[indicator].min()
            max_val = all_data[indicator].max()
            
            if max_val != min_val:
                phl_norm = (phl_latest[indicator] - min_val) / (max_val - min_val)
                sea_norm = (sea_avg[indicator] - min_val) / (max_val - min_val)
                
                normalized_phl.append(phl_norm)
                normalized_sea.append(sea_norm)
                categories.append(indicator.replace('_', ' ').title())
    
    N = len(categories)
    
    if N > 0:
        # Compute angles
        angles = [n / float(N) * 2 * pi for n in range(N)]
        normalized_phl += normalized_phl[:1]
        normalized_sea += normalized_sea[:1]
        angles += angles[:1]
        
        # Create radar chart
        fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))
        
        ax.plot(angles, normalized_phl, 'o-', linewidth=2, label='Philippines', color='#d62728')
        ax.fill(angles, normalized_phl, alpha=0.25, color='#d62728')
        ax.plot(angles, normalized_sea, 'o-', linewidth=2, label='ASEAN Average', color='#1f77b4')
        ax.fill(angles, normalized_sea, alpha=0.25, color='#1f77b4')
        
        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(categories, size=10)
        ax.set_ylim(0, 1)
        ax.set_title('Philippines vs ASEAN Average\nMulti-Dimensional Development Comparison', 
                    size=14, fontweight='bold', pad=20)
        ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
        ax.grid(True)
        
        plt.tight_layout()
        plt.show()
        
        print(f"✓ Radar chart created with {N} indicators (normalized 0-1 scale)")
        print(f"  All values normalized relative to global min-max range")
    else:
        print("⚠ Not enough indicators with data")
```

5. **Correlation ≠ Causation**: Our analysis identifies associations, not definitive causal relationships

7. **Comparative Challenges**: Philippines' archipelago geography differs from mainland SEA countries

### Scope Limitations:
8. **Poverty Line**: $2.15/day may not reflect local realities; Philippines uses higher national poverty line
9. **Quality of Life**: Monetary poverty doesn't capture health, education, dignity dimensions
10. **Political Economy**: Our analysis doesn't deeply examine political and institutional factors

## Final Remarks

### The Path Forward for Philippines

While the Philippines faces significant poverty challenges, this analysis demonstrates that **rapid poverty reduction is achievable**. Vietnam started from a higher poverty rate (50% in 1992) and achieved near-elimination (<2% in 2023) in just 30 years. The Philippines has stronger starting conditions in many ways (higher GDP per capita, more developed services sector, higher literacy).

### What Must Change:

**From Growth-Focused to Inclusive Growth:**
- Philippines needs to shift from maximizing GDP growth to maximizing **pro-poor growth**
- Economic growth alone is necessary but not sufficient

**From Metro Manila-Centric to Nationwide Development:**
- Decentralize opportunities and infrastructure investment
- Connect rural areas to markets through roads, electricity, internet

**From Enrollment to Learning:**
- Education policy must focus on quality and relevance, not just quantity
- Vocational skills and agricultural training are as important as academic education

### Meeting SDG 1 by 2030:

Based on current trajectories, Philippines is **at risk** of missing the SDG 1 target (ending extreme poverty by 2030). At the current rate of reduction (~0.3 percentage points/year), poverty would still be ~13% in 2030.

**What it would take to succeed:**
- **Accelerate to 1.5+ percentage points/year** (Vietnam's average rate)
- This requires: infrastructure investment, education quality improvement, job creation in manufacturing, and reducing inequality

### Hope from the Data:

The comparative analysis shows the Philippines doesn't face insurmountable barriers:
- SEA neighbors proved poverty reduction is possible
- Philippines has resources and human capital
- What's needed is **political will** and **smart policies** that learn from successful peers

### Final Thought:

**Poverty is not inevitable—it's a policy choice.** Vietnam, Malaysia, and Thailand chose policies that reduced poverty rapidly. The Philippines can do the same. The data shows the path; the question is whether policymakers will take it.
